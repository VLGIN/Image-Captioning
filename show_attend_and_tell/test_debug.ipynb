```python
with open('caption_dataset/VizWiz.json') as f:
  data = json.load(f)
```

```python
for i in range(len(data["images"])):
  for j in range(len(data["images"][i]["sentences"])):
    raw = data["images"][i]["sentences"][j]["tokens"]
    tokens = data["images"][i]["sentences"][j]["raw"]
    data["images"][i]["sentences"][j]["tokens"] = tokens
    data["images"][i]["sentences"][j]["raw"] = raw
```
